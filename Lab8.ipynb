## Análisis formal

### Bucle externo (i):
$$
i = \frac{n}{2}, \frac{n}{2} + 1, \ldots, n
$$

**Número de iteraciones:**
$$
\frac{n}{2} + 1 \approx O(n)
$$

### Bucle medio (j):
$$
j = 1, 2, 3, \ldots, \frac{n}{2}
$$

**Número de iteraciones:**
$$
\frac{n}{2} \approx O(n)
$$

### Bucle interno (k):
$$
k = 1, 2, 4, 8, \ldots, n
$$

**Número de iteraciones:**
$$
\log_2(n) \approx O(\log n)
$$

### Fórmula total
$$
T(n) = \sum_{i=1}^{n} \sum_{j=1}^{n} \sum_{k=1}^{\log n} 1
$$

$$
T(n) = n \cdot n \cdot \log n = O(n^2 \log n)

```python
import time
import matplotlib.pyplot as plt

def function1(n):
    counter = 0
    for i in range(n//2, n + 1):
        j = 1
        while j + n//2 <= n:
            k = 1
            while k <= n:
                counter += 1
                k *= 2
            j += 1

# Tamaños de entrada
sizes = [1, 10, 100, 1000, 10000]
times = []

for n in sizes:
    start = time.time()
    function1(n)
    end = time.time()
    elapsed_time = end - start
    times.append(elapsed_time)
    print(f"n={n}: {elapsed_time:.6f}s")

plt.figure(figsize=(10, 6))
plt.plot(sizes, times, marker="o", linestyle="-", color="green")
plt.xscale("log")  
plt.yscale("linear") 
plt.xlabel("Tamaño de entrada (n)")
plt.ylabel("Tiempo de ejecución (s)")
plt.title("Complejidad O(n² log n)")
plt.grid(True, which="both", linestyle="--", alpha=0.5)

# Etiquetar los puntos
for i, (n, t) in enumerate(zip(sizes, times)):
    plt.annotate(f'{t:.2e}s', (sizes[i], times[i]), 
                 xytext=(5, 5), textcoords='offset points', fontsize=8)

plt.show()
```

```python
import random
import matplotlib.pyplot as plt

def linear_search(arr, x):
    comparisons = 0
    for i in range(len(arr)):
        comparisons += 1
        if arr[i] == x:
            return i, comparisons
    return -1, comparisons

# Configuración de tamaños de prueba y probabilidades
sizes = [10, 100, 1000, 5000]
probabilidades = [0, 0.25, 0.5, 0.75, 1]

# Inicializar estructura para resultados
resultados = {p: [] for p in probabilidades}


for n in sizes:
    arr = list(range(1, n+1))  
    print(f"n = {n}, tamaño del arreglo: {len(arr)}")
    
    for p in probabilidades:
        repeticiones = 200
        total_comparaciones = 0
        
        for _ in range(repeticiones):
            if random.random() < p:
                x = random.choice(arr)
            else:
                x = -1
                
            _, comps = linear_search(arr, x)
            total_comparaciones += comps
        
        promedio = total_comparaciones / repeticiones
        resultados[p].append(promedio)
        print(f"  p={p}: promedio = {promedio:.2f}")

# Tabla de resultados
print("\n" + "="*60)
print("RESULTADOS COMPLETOS")
print("="*60)
print("n\tp=0.00\tp=0.25\tp=0.50\tp=0.75\tp=1.00")
print("-" * 60)
for i, n in enumerate(sizes):
    print(f"{n}", end="\t")
    for p in probabilidades:
        print(f"{resultados[p][i]:.2f}", end="\t")
    print()

# Graficar
plt.figure(figsize=(12, 8))
for p in probabilidades:
    plt.plot(sizes, resultados[p], marker="o", linewidth=2, markersize=8, label=f"p={p}")

plt.xlabel("Tamaño del arreglo (n)", fontsize=12)
plt.ylabel("Comparaciones promedio", fontsize=12)
plt.title("Búsqueda Lineal - Comparaciones promedio vs Tamaño del arreglo", fontsize=14)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)

# Añadir etiquetas a los puntos
for p in probabilidades:
    for i, (n, prom) in enumerate(zip(sizes, resultados[p])):
        plt.annotate(f'{prom:.1f}', (n, prom), 
                    xytext=(5, 5), textcoords='offset points', 
                    fontsize=8, alpha=0.7)

plt.tight_layout()
plt.show()
```


En el análisis contaremos principalmente comparaciones `A[i] == x`, que es la operación dominante para buscar.

---

### Mejor caso

**Situación:** El elemento buscado $x$ se encuentra en la primera posición $i = 1$.

**Conteo de comparaciones:**
La función hace 1 comparación (comprueba $A[1] == x$) y devuelve el índice.

**Costo:** $T_{\text{mejor}}(n) = 1$

**Notación asintótica:**
$T_{\text{mejor}}(n) = \Theta(1)$ o $O(1)$

**Explicación:** Independientemente de $n$, si el valor está en el primer elemento la búsqueda termina inmediatamente; por eso es constante.

---

### Peor caso

**Situación:** El elemento no está en el arreglo, o está en la última posición $i = n$.

**Conteo de comparaciones:**
Si está en la última posición: se hacen comparaciones en $i = 1, 2, \ldots, n$ → $n$ comparaciones y luego devuelve.

Si no está, el bucle también hace $n$ comparaciones y luego retorna -1.

**Costo:** $T_{\text{peor}}(n) = n$

**Notación asintótica:**
$T_{\text{peor}}(n) = \Theta(n)$ o $O(n)$

**Explicación:** en el peor caso hay que verificar todos los elementos; por eso crece linealmente con $n$.

---

### Caso promedio — Modelo A (elemento está y su posición es uniforme)

**Supuesto:** El elemento buscado está en el arreglo y su posición $k$ es equiprobable entre $1$ y $n$.

Si $k$ es la posición donde se encuentra, el número de comparaciones es $k$.

**Esperanza (valor esperado) de comparaciones:**
$$
E[T(n)] = \frac{1 + 2 + 3 + \cdots + n}{n} = \frac{\frac{n(n+1)}{2}}{n} = \frac{n+1}{2}
$$

**Notación asintótica:**
$T_{\text{promedio}}(n) = \Theta(n)$

**Explicación paso a paso:**
Cada posición $k$ (1..n) ocurre con probabilidad $1/n$.

Esperanza de comparaciones = suma de ($k \cdot P(k)$) sobre $k$.

Simplificando queda $(n+1)/2$, que es proporcional a $n$. Para grandes $n$, $(n+1)/2 \sim n/2$, por lo que en notación asintótica es $O(n)$.

---

### Caso promedio — Modelo B (elemento puede no estar: probabilidad $p$ de que esté)

**Supuesto más general:**
Existe una probabilidad $p$ ($0 \leq p \leq 1$) de que el elemento buscado esté en el arreglo.

Si está, su posición es uniforme entre 1 y $n$.

Con probabilidad $1-p$ el elemento no está y se hacen $n$ comparaciones.

**Cálculo del tiempo esperado:**
Si está: esperanza de comparaciones = $(n+1)/2$

Si no está: comparaciones = $n$

Entonces:
$$
E[T(n)] = p \cdot \frac{n+1}{2} + (1-p) \cdot n
$$

**Simplificando:**
$$
E[T(n)] = \frac{p(n+1)}{2} + n(1-p) = n\left(1 - \frac{p}{2}\right) + \frac{p}{2}
$$

**Notación asintótica:**
$E[T(n)] = \Theta(n)$

porque el coeficiente de $n$ es constante (depende de $p$).

**Interpretación:**
Si $p = 1$ (siempre está), recuperamos $(n+1)/2$

Si $p = 0$ (nunca está), recuperamos $n$

En cualquier caso razonable, la dependencia principal es lineal en $n$.

---

### Relación entre los tres casos (resumen)

- **Mejor caso:** $\Theta(1)$ — termina en la primera comparación
- **Caso promedio:** $\Theta(n)$ — en promedio necesita revisar $\sim n/2$ elementos si está; si puede no estar depende de $p$ pero sigue siendo $\Theta(n)$
- **Peor caso:** $\Theta(n)$ — revisar todos los elementos si no está o está al final

## Ejercicio 5

## Enunciado 1:
Si $f(n) = \Theta(g(n))$ y $g(n) = \Theta(h(n))$, entonces $h(n) = \Theta(f(n))$.

## Análisis:

$f(n) = \Theta(g(n))$ significa:
$\exists c_1, c_2 > 0, n_0$ tal que $c_1 g(n) \leq f(n) \leq c_2 g(n)$ para $n \geq n_0$.

$g(n) = \Theta(h(n))$ significa:
$\exists c_3, c_4 > 0, n_1$ tal que $c_3 h(n) \leq g(n) \leq c_4 h(n)$ para $n \geq n_1$.

**Combinando:**
- De $f(n) \leq c_2 g(n)$ y $g(n) \leq c_4 h(n)$ → $f(n) \leq c_2 c_4 h(n)$
- De $f(n) \geq c_1 g(n)$ y $g(n) \geq c_3 h(n)$ → $f(n) \geq c_1 c_3 h(n)$

**Por tanto:**
$$
c_1 c_3 h(n) \leq f(n) \leq c_2 c_4 h(n) \quad \text{para } n \geq \max(n_0, n_1)
$$

Eso es $h(n) = \Theta(f(n))$.

## Conclusión: Verdadero.

## Enunciado 2:
Si $f(n) = O(g(n))$ y $g(n) = O(h(n))$, entonces $h(n) = \Omega(f(n))$.

## Análisis:

$f(n) = O(g(n))$: 
$\exists c_1, n_0$ tal que $f(n) \leq c_1 g(n)$ para $n \geq n_0$.

$g(n) = O(h(n))$: 
$\exists c_2, n_1$ tal que $g(n) \leq c_2 h(n)$ para $n \geq n_1$.

**Combinando:**
$$
f(n) \leq c_1 g(n) \leq c_1 c_2 h(n) \quad \Rightarrow \quad f(n) \leq C h(n)
$$
con $C = c_1 c_2$.

Esto significa $h(n) \geq (1/C) f(n)$ para $n$ grande, que es precisamente $h(n) = \Omega(f(n))$.

## Conclusión: Verdadero.

```python
def A(n):
    atupla = tuple(range(0, n))  # O(n)
    S = set()
    for i in range(0, n):
        for j in range(i + 1, n):
            S.add(atupla[i:j])   # añade la tupla (i,...,j-1) al set S
```

## Análisis:

**Construcción de atupla:** 
$O(n)$, despreciable.

**Bucles anidados:**

Para cada $i$ desde $0$ hasta $n-1$,
$j$ va desde $i+1$ hasta $n-1$.

**Número de iteraciones:**
$$
\sum_{i=0}^{n-1} (n-1-i) = \sum_{k=1}^{n} (k-1) = \frac{n(n-1)}{2}
$$

Eso es $O(n^2)$ iteraciones.

**Operación S.add(atupla[i:j]):**

- `atupla[i:j]` crea una nueva tupla de longitud $j-i$
- El costo de crear esa rebanada es $O(j-i)$ en tiempo
- Además, `S.add()` debe calcular el hash de la tupla, que en el peor caso recorre todos los elementos de la tupla, también $O(j-i)$
- Y debe verificar si ya está en el set (coste esperado $O(1)$ si no hay muchas colisiones, pero el cálculo del hash es $O(\text{len})$)

**Suma de costos:**
El costo total es la suma sobre $i,j$ de $O(j-i)$.

Sea $k = j-i$, con $1 \leq k \leq n-1$.

Para un $k$ fijo,nos surge la sigueinte pregunta: ¿cuántos pares $(i,j)$ cumplen $j-i = k$?

$i$ puede ser $0, 1, \ldots, n-1-k$, es decir, $n-k$ pares.

Entonces:
$$
\sum_{i=0}^{n-1} \sum_{j=i+1}^{n-1} (j-i) = \sum_{k=1}^{n-1} k \cdot (n-k)
$$

$$
= n \sum_{k=1}^{n-1} k - \sum_{k=1}^{n-1} k^2
$$

$$
= n \cdot \frac{(n-1)n}{2} - \frac{(n-1)n(2n-1)}{6}
$$

$$
= \frac{n^2(n-1)}{2} - \frac{(n-1)n(2n-1)}{6}
$$

$$
= \frac{n(n-1)}{6} [3n - (2n-1)] = \frac{n(n-1)(n+1)}{6}
$$

Es decir, $\Theta(n^3)$.

## Conclusión:
El enunciado dice $f(n) = \Theta(n!)$, pero en realidad es $\Theta(n^3)$, así que es **Falso**.
