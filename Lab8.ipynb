# Ejercicio 1


## Análisis formal

### Bucle externo (i):
$$
i = \frac{n}{2}, \frac{n}{2} + 1, \ldots, n
$$

**Número de iteraciones:**
$$
\frac{n}{2} + 1 \approx O(n)
$$

### Bucle medio (j):
$$
j = 1, 2, 3, \ldots, \frac{n}{2}
$$

**Número de iteraciones:**
$$
\frac{n}{2} \approx O(n)
$$

### Bucle interno (k):
$$
k = 1, 2, 4, 8, \ldots, n
$$

**Número de iteraciones:**
$$
\log_2(n) \approx O(\log n)
$$

### Fórmula total
$$
T(n) = \sum_{i=1}^{n} \sum_{j=1}^{n} \sum_{k=1}^{\log n} 1
$$

$$
T(n) = n \cdot n \cdot \log n = O(n^2 \log n)

```python
import time
import matplotlib.pyplot as plt

def function1(n):
    counter = 0
    for i in range(n//2, n+1):
        j = 1
        while j + n//2 <= n:
            k = 1
            while k <= n:
                counter += 1
                k *= 2
            j += 1
    return counter

# Profiling
sizes = [1, 10, 100, 1000, 10000, 100000, 1000000]
#sizes = [1, 10, 100, 1000, 10000]

times = []

for n in sizes:
    start = time.time()
    function1(n)
    end = time.time()
    times.append(end - start)

# Tabla de resultados
print("n\tTiempo (s)")
for n, t in zip(sizes, times):
    print(f"{n}\t{t:.6f}")

# Gráfica
plt.plot(sizes, times, marker="o")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Tamaño de entrada (n)")
plt.ylabel("Tiempo de ejecución (s)")
plt.title("Ejercicio 1: O(n^2 log n)")
plt.show()
```

# Ejercicio 2

## Análisis formal

### Condición inicial:
$$
O(1)
$$

### Bucle externo (i):
$$
i = 1, 2, \ldots, n \Rightarrow O(n)
$$

### Bucle interno (j):
Aunque parece un bucle $O(n)$, el `break` interrumpe después de la primera iteración, entonces:

$$
O(1)
$$

### Fórmula total
$$
T(n) = \sum_{i=1}^{n} 1 = O(n)

```python
import time
import matplotlib.pyplot as plt

def function2(n):
    if n <= 1:
        return
    for i in range(1, n + 1):
        for j in range(1, n + 1):
            _ = "Sequence"
            break 

# Tamaños de entrada
sizes = [1, 10, 100, 1000, 10000, 100000, 1000000]
times = []

for n in sizes:
    start = time.time()
    function2(n)
    end = time.time()
    elapsed_time = end - start
    times.append(elapsed_time)
    print(f"n={n}: {elapsed_time:.6f}s")

plt.figure(figsize=(10, 6))
plt.plot(sizes, times, marker="o", linestyle="-", color="blue")
plt.xscale("log")  
plt.yscale("linear") 
plt.xlabel("Tamaño de entrada (n)")
plt.ylabel("Tiempo de ejecución (s)")
plt.title("Complejidad O(n) - Ejercicio 2")
plt.grid(True, which="both", linestyle="--", alpha=0.5)

# Etiquetar los puntos
for i, (n, t) in enumerate(zip(sizes, times)):
    plt.annotate(f'{t:.2e}s', (sizes[i], times[i]), 
                 xytext=(5, 5), textcoords='offset points', fontsize=8)

plt.show()
```

## Ejercicio 4


En el análisis contaremos principalmente comparaciones `A[i] == x`, que es la operación dominante para buscar.

---

### Mejor caso

**Situación:** El elemento buscado $x$ se encuentra en la primera posición $i = 1$.

**Conteo de comparaciones:**
La función hace 1 comparación (comprueba $A[1] == x$) y devuelve el índice.

**Costo:** $T_{\text{mejor}}(n) = 1$

**Notación asintótica:**
$T_{\text{mejor}}(n) = \Theta(1)$ o $O(1)$

**Explicación:** Independientemente de $n$, si el valor está en el primer elemento la búsqueda termina inmediatamente; por eso es constante.

---

### Peor caso

**Situación:** El elemento no está en el arreglo, o está en la última posición $i = n$.

**Conteo de comparaciones:**
Si está en la última posición: se hacen comparaciones en $i = 1, 2, \ldots, n$ → $n$ comparaciones y luego devuelve.

Si no está, el bucle también hace $n$ comparaciones y luego retorna -1.

**Costo:** $T_{\text{peor}}(n) = n$

**Notación asintótica:**
$T_{\text{peor}}(n) = \Theta(n)$ o $O(n)$

**Explicación:** en el peor caso hay que verificar todos los elementos; por eso crece linealmente con $n$.

---

### Caso promedio — Modelo A (elemento está y su posición es uniforme)

**Supuesto:** El elemento buscado está en el arreglo y su posición $k$ es equiprobable entre $1$ y $n$.

Si $k$ es la posición donde se encuentra, el número de comparaciones es $k$.

**Esperanza (valor esperado) de comparaciones:**
$$
E[T(n)] = \frac{1 + 2 + 3 + \cdots + n}{n} = \frac{\frac{n(n+1)}{2}}{n} = \frac{n+1}{2}
$$

**Notación asintótica:**
$T_{\text{promedio}}(n) = \Theta(n)$

**Explicación paso a paso:**
Cada posición $k$ (1..n) ocurre con probabilidad $1/n$.

Esperanza de comparaciones = suma de ($k \cdot P(k)$) sobre $k$.

Simplificando queda $(n+1)/2$, que es proporcional a $n$. Para grandes $n$, $(n+1)/2 \sim n/2$, por lo que en notación asintótica es $O(n)$.

---

### Caso promedio — Modelo B (elemento puede no estar: probabilidad $p$ de que esté)

**Supuesto más general:**
Existe una probabilidad $p$ ($0 \leq p \leq 1$) de que el elemento buscado esté en el arreglo.

Si está, su posición es uniforme entre 1 y $n$.

Con probabilidad $1-p$ el elemento no está y se hacen $n$ comparaciones.

**Cálculo del tiempo esperado:**
Si está: esperanza de comparaciones = $(n+1)/2$

Si no está: comparaciones = $n$

Entonces:
$$
E[T(n)] = p \cdot \frac{n+1}{2} + (1-p) \cdot n
$$

**Simplificando:**
$$
E[T(n)] = \frac{p(n+1)}{2} + n(1-p) = n\left(1 - \frac{p}{2}\right) + \frac{p}{2}
$$

**Notación asintótica:**
$E[T(n)] = \Theta(n)$

porque el coeficiente de $n$ es constante (depende de $p$).

**Interpretación:**
Si $p = 1$ (siempre está), recuperamos $(n+1)/2$

Si $p = 0$ (nunca está), recuperamos $n$

En cualquier caso razonable, la dependencia principal es lineal en $n$.

---

### Relación entre los tres casos (resumen)

- **Mejor caso:** $\Theta(1)$ — termina en la primera comparación
- **Caso promedio:** $\Theta(n)$ — en promedio necesita revisar $\sim n/2$ elementos si está; si puede no estar depende de $p$ pero sigue siendo $\Theta(n)$
- **Peor caso:** $\Theta(n)$ — revisar todos los elementos si no está o está al final
