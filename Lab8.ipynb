# Ejercicio 1


## Análisis formal

### Bucle externo (i):
$$
i = \frac{n}{2}, \frac{n}{2} + 1, \ldots, n
$$

**Número de iteraciones:**
$$
\frac{n}{2} + 1 \approx O(n)
$$

### Bucle medio (j):
$$
j = 1, 2, 3, \ldots, \frac{n}{2}
$$

**Número de iteraciones:**
$$
\frac{n}{2} \approx O(n)
$$

### Bucle interno (k):
$$
k = 1, 2, 4, 8, \ldots, n
$$

**Número de iteraciones:**
$$
\log_2(n) \approx O(\log n)
$$

### Fórmula total
$$
T(n) = \sum_{i=1}^{n} \sum_{j=1}^{n} \sum_{k=1}^{\log n} 1
$$

$$
T(n) = n \cdot n \cdot \log n = O(n^2 \log n)

```python
import time
import matplotlib.pyplot as plt

def function1(n):
    counter = 0
    for i in range(n//2, n+1):
        j = 1
        while j + n//2 <= n:
            k = 1
            while k <= n:
                counter += 1
                k *= 2
            j += 1
    return counter

# Profiling
#sizes = [1, 10, 100, 1000, 10000, 100000, 1000000]
sizes = [1, 10, 100, 1000, 10000]

times = []

for n in sizes:
    start = time.time()
    function1(n)
    end = time.time()
    times.append(end - start)

# Tabla de resultados
print("n\tTiempo (s)")
for n, t in zip(sizes, times):
    print(f"{n}\t{t:.6f}")

# Gráfica
plt.plot(sizes, times, marker="o")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Tamaño de entrada (n)")
plt.ylabel("Tiempo de ejecución (s)")
plt.title("Ejercicio 1: O(n^2 log n)")
plt.show()
```

# Ejercicio 2

## Análisis formal

### Condición inicial:
$$
O(1)
$$

### Bucle externo (i):
$$
i = 1, 2, \ldots, n \Rightarrow O(n)
$$

### Bucle interno (j):
Aunque parece un bucle $O(n)$, el `break` interrumpe después de la primera iteración, entonces:

$$
O(1)
$$

### Fórmula total
$$
T(n) = \sum_{i=1}^{n} 1 = O(n)

```python
def function2(n):
    if n <= 1:
        return
    for i in range(1, n+1):
        for j in range(1, n+1):
            _ = "Sequence"  
            break


sizes = [1, 10, 100, 1000, 10000, 100000, 1000000]
times = []

for n in sizes:
    start = time.time()
    function2(n)
    end = time.time()
    times.append(end - start)

# Tabla de resultados
print("n\tTiempo (s)")
for n, t in zip(sizes, times):
    print(f"{n}\t{t:.6f}")

# Gráfica
plt.plot(sizes, times, marker="o")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Tamaño de entrada (n)")
plt.ylabel("Tiempo de ejecución (s)")
plt.title("Ejercicio 2: O(n)")
plt.show()
```
